```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Specify the path to your CSV file
file_path = r'R:\ML\ML nd Dl by codebasics\csv\income.csv'

# Load the CSV file into a DataFrame
df= pd.read_csv(file_path)
df
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df1
```

```python
km.cluster_centers_ #centroid for each cluster with respect xaxis  and y axis
#array([[3.29090909e+01, 5.61363636e+04],  ==>centroid for 1st cluster with respect xaxis  and y axis since draw scatter plot along x-axis and y axis so two points for each cluster
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age'] 
```

```python

plt.scatter(df.Age,df['Income($)'])
```

```python
#creating new object of k means cluster through this obejct predict
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python

sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)#creating new object of k means cluster through this obejct predict
    km.fit(df[['Age','Income($)']])   
    sse.append(km.inertia_)
sse    
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng, sse, marker='o')  # 'o' is for circle markers
```

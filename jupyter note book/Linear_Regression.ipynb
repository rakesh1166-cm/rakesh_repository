```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_home and predictions5 are already defined as per your code
df_home['predicted_price'] = predictions5

# Create a figure with subplots for seaborn plots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Bar plot for actual vs predicted prices
sns.barplot(x=df_home.index, y=df_home['price'], color='red', alpha=0.6, label='Actual', ax=axes[0, 0])
sns.barplot(x=df_home.index, y=df_home['predicted_price'], color='blue', alpha=0.6, label='Predicted', ax=axes[0, 0])
axes[0, 0].set_xlabel('Index')
axes[0, 0].set_ylabel('Price')
axes[0, 0].set_title('Actual vs Predicted Prices')
axes[0, 0].legend()

# Distribution plot for actual prices
sns.distplot(df_home['price'], color='red', kde=True, ax=axes[0, 1])
axes[0, 1].set_xlabel('Price')
axes[0, 1].set_ylabel('Density')
axes[0, 1].set_title('Distribution of Actual Prices')

# Distribution plot for predicted prices
sns.distplot(df_home['predicted_price'], color='blue', kde=True, ax=axes[1, 0])
axes[1, 0].set_xlabel('Predicted Price')
axes[1, 0].set_ylabel('Density')
axes[1, 0].set_title('Distribution of Predicted Prices')

# KDE plot for actual vs predicted prices
sns.kdeplot(df_home['price'], color='red', label='Actual', fill=True, ax=axes[1, 1])
sns.kdeplot(df_home['predicted_price'], color='blue', label='Predicted', fill=True, ax=axes[1, 1])
axes[1, 1].set_xlabel('Price')
axes[1, 1].set_ylabel('Density')
axes[1, 1].set_title('KDE Plot of Actual vs Predicted Prices')
axes[1, 1].legend()

# Adjust layout for 2x2 grid
plt.tight_layout()

# Separate figure for the joint plot to avoid layout issues
plt.figure(figsize=(8, 8))
sns.jointplot(x=df_home['price'], y=df_home['predicted_price'], kind='reg', color='purple')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.suptitle('Joint Plot of Actual vs Predicted Prices', y=1.02)

# Show all plots
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_home and predictions5 are already defined as per your code
df_home['predicted_price'] = predictions5

# Define plot details
plot_titles = [
    'Actual vs Predicted Prices',
    'Distribution of Actual Prices',
    'Distribution of Predicted Prices',
    'KDE Plot of Actual vs Predicted Prices',
    'Joint Plot of Actual vs Predicted Prices'
]
plot_functions = [
    'bar', 'dist', 'dist', 'kde', 'joint'
]
plot_colors = [
    ('red', 'blue'), 'red', 'blue', ('red', 'blue'), 'purple'
]

# Create a figure with subplots for seaborn plots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Loop through the plot details and create the plots
for i, (title, func, color) in enumerate(zip(plot_titles, plot_functions, plot_colors)):
    if i < 4:  # First 4 plots go to the 2x2 grid
        ax = axes[i // 2, i % 2]
        if func == 'bar':
            sns.barplot(x=df_home.index, y=df_home['price'], color=color[0], alpha=0.6, label='Actual', ax=ax)
            sns.barplot(x=df_home.index, y=df_home['predicted_price'], color=color[1], alpha=0.6, label='Predicted', ax=ax)
            ax.set_xlabel('Index')
            ax.set_ylabel('Price')
            ax.legend()
        elif func == 'dist':
            sns.distplot(df_home['price'] if color == 'red' else df_home['predicted_price'], color=color, kde=True, ax=ax)
            ax.set_xlabel('Price')
            ax.set_ylabel('Density')
        elif func == 'kde':
            sns.kdeplot(df_home['price'], color=color[0], label='Actual', fill=True, ax=ax)
            sns.kdeplot(df_home['predicted_price'], color=color[1], label='Predicted', fill=True, ax=ax)
            ax.set_xlabel('Price')
            ax.set_ylabel('Density')
            ax.legend()
        ax.set_title(title)
    else:  # Last plot goes to a new figure
        sns.jointplot(x=df_home['price'], y=df_home['predicted_price'], kind='reg', color=color)
        plt.xlabel('Actual Price')
        plt.ylabel('Predicted Price')
        plt.suptitle(title, y=1.02)

# Adjust layout for the 2x2 grid
plt.tight_layout()

# Show all plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
#         print (m_curr,b_curr, i)
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
     
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```

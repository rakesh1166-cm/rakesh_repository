# Write a function to display range of image from the training dataset X_train

```python
def plot_samples(indices):
    n = len(indices)  # Number of images to plot
    cols = 2  # Number of columns you want
    rows = (n + cols - 1) // cols  # Calculate the number of rows needed

    plt.figure(figsize=(15, rows * 5))  # Adjust figure size accordingly
    for i, index in enumerate(indices):
        plt.subplot(rows, cols, i + 1)  # Create subplots dynamically
        plt.imshow(X_train[index])
        plt.title(f"Class: {y_train[index][0]}")
        plt.axis('off')  # Hide axes
    plt.show()
```

```python
plot_samples([0, 1, 2, 3, 4])
```

# Preprocessing: Scale images

```python
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
X_train_scaled
```

```python
X_test_scaled
```

```python
y_train
```

```python
from keras.utils import to_categorical

# Assuming y_train and y_test are your labels
y_train_categorical = to_categorical(y_train, num_classes=10)
y_test_categorical = to_categorical(y_test, num_classes=10)
y_train_categorical
```

```python
y_train[0:5]
```

```python
y_train_categorical[0:5]
```

# Model building and training

```python
model = keras.Sequential([
        keras.layers.Flatten(input_shape=(32,32,3)),
        keras.layers.Dense(3000, activation='relu'),
        keras.layers.Dense(1000, activation='relu'),
        keras.layers.Dense(10, activation='sigmoid')    
    ])

model.compile(optimizer='SGD',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

```python
def get_model():
    model = keras.Sequential([
            keras.layers.Flatten(input_shape=(32,32,3)),
            keras.layers.Dense(3000, activation='relu'),
            keras.layers.Dense(1000, activation='relu'),
            keras.layers.Dense(10, activation='sigmoid')    
        ])

    model.compile(optimizer='SGD',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

# Measure training time on a CPU

```python

%%timeit -n1 -r1 
with tf.device('/CPU:0'):
    cpu_model = get_model()
    cpu_model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

# Lets measure training time on a GPU (I've NVIDIA Titan RTX)

```python
%%timeit -n1 -r1 
with tf.device('/GPU:0'):
    cpu_model = get_model()
    cpu_model.fit(X_train_scaled, y_train_categorical, epochs=1)
```

# Lets run same test for 10 epocs

```python
%%timeit -n1 -r1 
with tf.device('/CPU:0'):
    cpu_model = get_model()
    cpu_model.fit(X_train_scaled, y_train_categorical, epochs=10)
```

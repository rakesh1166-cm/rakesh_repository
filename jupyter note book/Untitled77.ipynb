```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pathlib

# Step 1: Point to the dataset directory
data_dir = pathlib.Path(r'R:\ML\DEEP learning\csv\datasets\flower_photos')

# Step 2: Load an example image from the 'roses' folder
rose_images = list(data_dir.glob('roses/*'))
image_path = rose_images[0]  # Load the first image from 'roses'

# Step 3: Load and decode the image
img = tf.io.read_file(str(image_path))
img = tf.image.decode_jpeg(img)

# Step 4: Display the original image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()

# Step 5: Define data augmentation pipeline
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomContrast(0.2),
    tf.keras.layers.RandomBrightness(factor=0.2),
    tf.keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1),
    tf.keras.layers.RandomCrop(180, 180)
])

# Step 6: Apply augmentation and visualize multiple augmented images
plt.figure(figsize=(10, 10))
for i in range(9):  # Generate 9 augmented images
    augmented_image = data_augmentation(tf.expand_dims(img, 0))  # Expand dims for batch
    plt.subplot(3, 3, i + 1)
    plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))  # Remove batch dim for display
    plt.axis('off')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pathlib

# Load a specific image (e.g., from the 'roses' folder of the flower_photos dataset)
data_dir = pathlib.Path(r'R:\ML\DEEP learning\csv\datasets\flower_photos')
rose_images = list(data_dir.glob('roses/*'))  # List all images in 'roses'
image_path = rose_images[0]  # Pick the first rose image

# Read and decode the image
img = tf.io.read_file(str(image_path))
img = tf.image.decode_jpeg(img)

# Display the original image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()

# Define a single data augmentation: RandomRotation by 20%
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomRotation(0.2)  # Rotate by a random factor (0.2 means 20% of 360 degrees)
])

# Apply the augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))  # Expand dims to include batch size

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Rotation")
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pathlib

# Load the image
data_dir = pathlib.Path(r'R:\ML\DEEP learning\csv\datasets\flower_photos')
rose_images = list(data_dir.glob('roses/*'))
image_path = rose_images[0]
img = tf.io.read_file(str(image_path))
img = tf.image.decode_jpeg(img)

# Display the original image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()

# Define random rotation augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomRotation(0.2)  # Rotate randomly by 20%
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Rotation")
plt.show()
```

```python
# Define random zoom augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomZoom(0.1)  # Zoom in by 10% randomly
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Zoom")
plt.show()
```

```python
# Define random brightness adjustment augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomBrightness(factor=0.2)  # Adjust brightness by 20%
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Brightness Adjustment")
plt.show()
```

```python
# Define random contrast adjustment augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomContrast(0.2)  # Adjust contrast by 20%
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Contrast Adjustment")
plt.show()
```

```python
# Define random zoom and center cropping augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomZoom(0.1),  # Zoom in by 10% randomly
    tf.keras.layers.CenterCrop(height=180, width=180)  # Crop the center to 180x180
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Zoom and Cropping")
plt.show()
```

```python
# Define a combination of random flip, rotation, zoom, and contrast adjustment
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),   # Randomly flip horizontally
    tf.keras.layers.RandomRotation(0.2),        # Rotate randomly by 20%
    tf.keras.layers.RandomZoom(0.2),            # Zoom in/out by 20%
    tf.keras.layers.RandomContrast(0.2)         # Adjust contrast by 20%
])

# Apply augmentation
augmented_image = data_augmentation(tf.expand_dims(img, 0))

# Display the augmented image
plt.imshow(tf.squeeze(augmented_image).numpy().astype("uint8"))
plt.axis('off')
plt.title("Augmented Image - Random Flip, Rotation, and Zoom Combined")
plt.show()
```

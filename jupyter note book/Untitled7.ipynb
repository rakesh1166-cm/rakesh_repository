```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# Point Markers
y1 = np.sin(x)
plt.plot(x, y1, '.', label='point')

y2 = np.cos(x)
plt.plot(x, y2, ',', label='pixel')

y3 = np.sin(x) + 1
plt.plot(x, y3, 'o', label='circle')

y4 = np.cos(x) + 1
plt.plot(x, y4, 'v', label='triangle_down')

y5 = np.sin(x) + 2
plt.plot(x, y5, '^', label='triangle_up')

y6 = np.cos(x) + 2
plt.plot(x, y6, '<', label='triangle_left')

y7 = np.sin(x) + 3
plt.plot(x, y7, '>', label='triangle_right')



plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)



# Line Markers
y8 = np.sin(x) + 4
plt.plot(x, y8, '-', label='solid')

y9 = np.cos(x) + 4
plt.plot(x, y9, '--', label='dashed')

y10 = np.sin(x) + 5
plt.plot(x, y10, '-.', label='dash-dot')

y11 = np.cos(x) + 5
plt.plot(x, y11, ':', label='dotted')



plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)



# Combination Markers
y12 = np.sin(x) + 6
plt.plot(x, y12, 'ro', label='red circle')

y13 = np.cos(x) + 6
plt.plot(x, y13, 'bo', label='blue circle')

y14 = np.sin(x) + 7
plt.plot(x, y14, 'go', label='green circle')

y15 = np.cos(x) + 7
plt.plot(x, y15, 'yo', label='yellow circle')

y16 = np.sin(x) + 8
plt.plot(x, y16, 'ko', label='black circle')

y17 = np.cos(x) + 8
plt.plot(x, y17, 'rx', label='red x-mark')

y18 = np.sin(x) + 9
plt.plot(x, y18, 'bx', label='blue x-mark')



plt.legend()
plt.show()
```

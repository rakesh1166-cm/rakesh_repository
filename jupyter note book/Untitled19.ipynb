```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Specify the path to your CSV file
file_path = r'R:\ML\ML nd Dl by codebasics\csv\insurance_data.csv'

# Load the CSV file into a DataFrame
df= pd.read_csv(file_path)
df
```

```python
df=df.sort_values('age')
df   
```

```python
print(df.describe())

# Histogram of ages
plt.figure(figsize=(10, 5))
plt.hist(df['age'], bins=10, color='blue', alpha=0.7, rwidth=0.85)
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Bar chart for insurance purchase counts
insurance_counts = df['bought_insurance'].value_counts()
plt.figure(figsize=(7, 5))
insurance_counts.plot(kind='bar', color=['red', 'green'])
plt.title('Insurance Purchase Counts')
plt.xlabel('Bought Insurance (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 5))
plt.scatter(df['age'], df['bought_insurance'], color='purple', alpha=0.5)
plt.title('Scatter Plot of Age vs. Bought Insurance')
plt.xlabel('Age')
plt.ylabel('Bought Insurance (1 = Yes, 0 = No)')
plt.grid(True)
plt.yticks([0, 1], ['No', 'Yes'])  # Just to make it clear that these are categorical labels
plt.show()
```

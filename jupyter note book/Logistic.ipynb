```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Specify the path to your CSV file
file_path = r'R:\ML\ML nd Dl by codebasics\csv\insurance_data.csv'

# Load the CSV file into a DataFrame
df= pd.read_csv(file_path)
df
```

```python
pd.crosstab(df.age,df.bought_insurance).plot(kind='bar')
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
ct = pd.crosstab(df.age, df.bought_insurance)

# Flatten the crosstab to use for a scatter plot
# This will repeat the 'age' index based on the counts for plotting
x, y, sizes = [], [], []
for age in ct.index:
    for bought in ct.columns:
        x.append(age)  # Age
        y.append(bought)  # Bought insurance
        sizes.append(ct.loc[age, bought])  # Frequency as size of the dot
print(sizes)    
print(x)
print(y)
plt.scatter(x, y, s=sizes)  # 's' controls the size of each point
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Scatter Plot of Insurance Purchase by Age')
plt.xticks(ct.index)  # Only show ages present in the crosstab
plt.yticks(ct.columns)  # Only show 0 and 1 for bought_insurance
plt.grid(True)
plt.show()
```

```python
print(df.describe())

# Histogram of ages
plt.figure(figsize=(10, 5))
plt.hist(df['age'], bins=10, color='blue', alpha=0.7, rwidth=0.85)
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Bar chart for insurance purchase counts
insurance_counts = df['bought_insurance'].value_counts()
plt.figure(figsize=(7, 5))
insurance_counts.plot(kind='bar', color=['red', 'green'])
plt.title('Insurance Purchase Counts')
plt.xlabel('Bought Insurance (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 5))
plt.scatter(df['age'], df['bought_insurance'], color='purple', alpha=0.5)
plt.title('Scatter Plot of Age vs. Bought Insurance')
plt.xlabel('Age')
plt.ylabel('Bought Insurance (1 = Yes, 0 = No)')
plt.grid(True)
plt.yticks([0, 1], ['No', 'Yes'])  # Just to make it clear that these are categorical labels
plt.show()
```

```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def prediction_function(age):
    z = 0.042 * age - 1.53 # 0.04150133 ~ 0.042 and -1.52726963 ~ -1.53
    y = sigmoid(z)
    return y
```

```python
age = 35
prediction_function(age)
#0.485 is less than 0.5 which means person with 35 age will not buy insurance
```

```python
age = 43
prediction_function(age)
#0.485 is more than 0.5 which means person with 43 will buy the insurance
```

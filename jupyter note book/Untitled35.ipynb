```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# Example DataFrame
df = pd.DataFrame({
    'Age': [29, 30, 34, 40, 41],
    'Income($)': [60000, 80000, 120000, 150000, 180000],
    'Credit Score': [650, 700, 750, 800, 850]
})

# Scatter plot with Age, Income, and Credit Score as color
plt.scatter(df['Age'], df['Income($)'], c=df['Credit Score'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.colorbar(label='Credit Score')  # Show color scale
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Specify the path to your CSV file
file_path = r'R:\ML\ML nd Dl by codebasics\csv\salaries.csv'

# Load the CSV file into a DataFrame
df= pd.read_csv(file_path)
new_df=df
df
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
```

```python
target = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
#df.Sex = df.Sex.map({'male': 1, 'female': 2})
##df['Sex'] = df['Sex'].apply(lambda x: 1 if x == 'male' else 2)
```

```python
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
#to know whata are the parameter in kernal  put cursor DecisionTreeClassifier() and click shift+tab
```

```python
model.fit(inputs_n, target)
```

```python

model.score(inputs_n,target)
```

```python
# Specify the path to your CSV file
file_path = r'R:\ML\ML nd Dl by codebasics\csv\salaries.csv'

# Load the CSV file into a DataFrame
my_df= pd.read_csv(file_path)
```

```python
# Create a mapping from the factorize method
mapping, _ = pd.factorize(my_df['company'])
print(mapping)
# Convert 'company' to numeric in a single line using apply and the previously created mapping
my_df['company'] = my_df['company'].apply(lambda x: mapping[my_df['company'].tolist().index(x)])

print(my_df)
```

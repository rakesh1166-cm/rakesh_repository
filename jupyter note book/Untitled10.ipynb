```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: Daily closing prices of multiple stocks over a month
data = {
    'Date': pd.date_range(start='2023-01-01', periods=30, freq='D'),
    'Stock_A': [150 + i*0.5 for i in range(30)],
    'Stock_B': [100 + i*0.3 for i in range(30)],
    'Stock_C': [200 + i*0.2 for i in range(30)]
}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
for stock in df:
    print(df[stock])
```

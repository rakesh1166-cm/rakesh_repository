The activation function adds "non-linearity," allowing the network to make sense of more diverse and sophisticated data by mimicking how the brain works, making decisions based on the input it receives.
 
 
 Enable Learning: Activation functions decide whether a neuron should be activated (i.e., whether the information should pass through to the next layer). This helps the network learn meaningful relationships from the data.

Control Output: Different activation functions produce different kinds of outputs. For example, some may squash the output to a range between 0 and 1 (like the sigmoid function), or allow negative and positive values (like the tanh function), while others (like ReLU) only allow positive outputs. This affects how the network behaves and adapts to learning tasks.

Prevent Vanishing Gradient: Some activation functions, like ReLU, help prevent the "vanishing gradient" problem, where gradients become too small for the network to learn effectively during backpropagation

```python
#Sigmoid==Range: (0, 1)
```

```python
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

```python
sigmoid(0.5)
```

```python
#tanh==>Range: (-1, 1)
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

```python
#ReLU  ==>Range: [0, ∞)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-100)
```

```python

relu(8)
```

```python
relu(-15)
```

```python
#Leaky ReLU  ==>Range: (-∞, ∞)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(8)
```

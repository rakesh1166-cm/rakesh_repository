```python
#Impact of salary on employee retention
pd.crosstab(df.salary,df.left).plot(kind='bar')
```

```python
pd.crosstab(df.Department,df.left).plot(kind='bar')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 15))  # Adjust figure size as necessary
ax = axes.flatten()  # Flatten the axes array for easy iteration

# Counter for the subplot index
index = 0

# Loop through each column in the DataFrame except 'left'
for column in df.columns:
    if column != 'left':  # Skip the 'left' column to avoid self-comparison
        # Generate the crosstab
        ct = pd.crosstab(df[column], df['left'])

        # Plot the crosstab on specified subplot axis
        ct.plot(kind='bar', ax=ax[index], title=f'Count of Left by {column}')
        ax[index].set_xlabel(column)  # Set x-label as the current column name
        ax[index].set_ylabel('Count')  # Set y-label as 'Count'
        ax[index].set_xticklabels(ax[index].get_xticklabels(), rotation=0)  # Manage x-tick labels

        # Increment subplot index
        index += 1

# Adjust layout to prevent overlap
plt.tight_layout()

# Display all plots
plt.show()
```

```python
crosstab_result = pd.crosstab(grouped_means['left'], grouped_means['promotion_last_5years'])
crosstab_result
```

```python
columns_to_process = grouped_means.columns.difference(['left'])

# Set up the figure for subplots
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(10, 10))  # Adjust the figure size to your liking
axes = axes.flatten()  # Flatten the axes array for easy indexing

for idx, col in enumerate(columns_to_process):
    if idx < 16:  # We have space for only 16 plots
        # Data preparation for plotting
        heights = grouped_means[col].values  # Get the column values for heights
        x_labels = grouped_means['left'].values  # Corresponding 'left' values
        
        # Plotting the bar plot in a subplot
        axes[idx].bar(x_labels, heights, color=['blue', 'red'])
        axes[idx].set_title(f'{col}')
        axes[idx].set_xlabel('Left')
        axes[idx].set_xticks([0, 1])
        axes[idx].set_xticklabels(['0', '1'])
        axes[idx].set_ylabel('Value')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
subdf = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
subdf.head()

```

```python
salary_dummies = pd.get_dummies(subdf.salary, prefix="salary")
salary_dummies = salary_dummies.astype(int)
```

```python
df_with_dummies = pd.concat([subdf,salary_dummies],axis='columns')
```

```python
df_with_dummies.head()
```

```python
df_with_dummies.drop('salary',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
y = df.left
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python

model.predict(X_test)
```

```python
#Accuracy of the model
```

```python
model.score(X_test,y_test)
```

```python
print("Column names:", X_test)
y_predicted = model.predict(X_test)
print("Column names:", y_predicted)
# Count the predicted results
predicted_counts = pd.Series(y_predicted).value_counts()
print("Column names:", predicted_counts)
# Plot the bar chart for predicted data
plt.figure(figsize=(7, 5))
predicted_counts.sort_index().plot(kind='bar', color=['red', 'green'])  # Adjust colors if needed
plt.title('Predicted left  Counts')
plt.xlabel('left company (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
#True Positives (TP): The model correctly predicted 7581 instances as positive.
#False Positives (FP): The model incorrectly predicted 419 instances as positive when they were actually negative.
#False Negatives (FN): The model incorrectly predicted 1851 instances as negative when they were actually positive.
#True Negatives (TN): The model correctly predicted 649 instances as negative.
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

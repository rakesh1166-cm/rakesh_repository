```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(10, 4))

# Subplot 1
plt.subplot(1, 2, 2)
plt.plot(x, y1, label='sin(x)')
plt.title('Subplot 1')
plt.legend()

# Subplot 2
plt.subplot(1, 2, 1)
plt.plot(x, y2, label='cos(x)')
plt.title('Subplot 2')
plt.legend()

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
